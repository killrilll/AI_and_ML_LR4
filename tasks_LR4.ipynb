## 8. Арифметические операции с Series.

Создайте два Series:

1. s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
2. s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

Выполните сложение s1 + s2 . Объясните, почему в результате появляются NaN, и замените их на 0.

```python
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

result = s1 + s2
result_filled = result.fillna(0)

print(f"Результат сложения:\n{result}")
print(f"Замена NaN на нули:\n{result_filled}")
```

**Объяснение:** Если индексы не совпадают, то в соответствующих местах появятся NaN. Индекс 'a' есть только в s1, а индекс 'e' есть только в s2, поэтому при сложении эти места заполняются NaN.

## 10. Основные статистические методы.

Создайте Series из 20 случайных чисел от 50 до 150 (используйте np.random.randint ). Найдите сумму, среднее, минимальное и максимальное значение. Выведите также стандартное отклонение.

## 15. Построение графика на основе Series.

Создайте Series , где индексами будут даты с 1 по 30 марта 2024 года, а
значениями – случайные числа от 50 до 150. Постройте график значений с
помощью matplotlib. Добавьте заголовок, подписи осей и сетку.

```python
s1 = pd.Series(np.random.randint(50, 151, 30), pd.date_range(start="2024-03-01", periods=30, freq="D"))

plt.figure(figsize=(10, 8))
plt.plot(s1.index, s1.values, marker="o", linestyle="-", color="black", label="Цена")

plt.title("Изменение значений за март 2024 года")
plt.xlabel("Дата")
plt.ylabel("Цена")
plt.grid(True)
plt.legend()

plt.show()
```

## Динамика популярности запроса в поисковике.

Создайте Series, где индексами будут даты за последние 30 дней (pd.date_range('2025-03-03', periods=30, freq='D')), а значениями – случайные значения популярности запроса (от 0 до 100). Найдите даты, когда популярность превышала 80, и выделите эти точки на графике маркерами.

```python
s1 = pd.Series(np.random.randint(0, 101, 30), pd.date_range('2025-03-01', periods=30, freq='D'))

high_date_s1 = s1[s1 > 80]

plt.figure(figsize=(10, 6))
plt.plot(s1.index, s1.values, marker="o", linestyle="-", color="black", label="Популярность")

plt.scatter(high_date_s1.index, high_date_s1.values, color="r", label="> 80", zorder=2)

plt.title("Популярность запроса за последние 30 дней")
plt.xlabel("Дата")
plt.ylabel("Популярность")
plt.grid()
plt.legend()

plt.show()
```
